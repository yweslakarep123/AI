<h1>CLEANING RATE COLUMN

```python
df['rate'].head()
```

```python
df['rate'].unique()
```

```python
def clean_rate(val):
  if(val=='NEW' or val=='-'):
    return np.nan
  else:
    val = str(val).split('/')
    val = val[0]
    return float(val)

df['rate'] = df['rate'].apply(clean_rate)
df['rate'].head()
```

```python
df['rate'].unique()
```

<h1>CLEAN COST2PLATES

```python
df['cost2plate'].unique()
```

```python
def clean_cost(val):
  val = str(val)
  if ',' in val:
    val = val.replace(',','')
    return float(val)
  else:
    return float(val)

df['cost2plate'] = df['cost2plate'].apply(clean_cost)
df['cost2plate'].unique()
```

```python
df.head()
```

```python
df = df.drop(['menu_item'],axis=1)
df.head()
```

<h1>CLEAN REST_TYPE</H1>

```python
df['rest_type'].unique()
```

```python
a = df['rest_type'].value_counts(ascending=False)
a[:5]
```

```python
rest_type = df['rest_type'].value_counts()
rest_type_1000 = rest_type[rest_type<1000]
rest_type_1000
```

```python
def clean_rest(val):
  if val in rest_type_1000:
    return 'others'
  else:
    return val


df['rest_type'] = df['rest_type'].apply(clean_rest)
df['rest_type'].value_counts()
```

```python
df.info()
```

```python
location = df['location'].value_counts()
location_300 = location[location<300]
location_300
```

```python
def clean_loc(val):
  if val in location_300:
    return 'others'
  else:
    return val

df['location'] = df['location'].apply(clean_loc)
df['location'].value_counts()
```

<h1>CLEAN COISINES</h1>

```python
df['cuisines'].value_counts()
```

```python
a = df['cuisines'].value_counts()
a[93//4:93//2]
```

```python
cuisines = df['cuisines'].value_counts()
cuisines_100 = cuisines[cuisines<100]
cuisines_100
```

```python
def clean_cui(val):
  if val in cuisines_100:
    return 'others'
  else:
    return val

df['cuisines'] = df['cuisines'].apply(clean_cui)
df['cuisines'].value_counts()
```

<h2>LOCATION AND VOTES</h2>
<p> kalau saya ingin buka di lokasi ini maka yang optimal dimana jika dilihar dari votes?</h2>

```python
df_loc_vo_1 = df[['location','votes']]
# df_loc_vo_1 = df_loc_vo_1.drop_duplicates()

df_loc_vo_2 = df_loc_vo_1.groupby(['location'])['votes'].sum()
df_loc_vo_2 = df_loc_vo_2.to_frame()
df_loc_vo_2 = df_loc_vo_2.sort_values('votes',ascending=False)
df_loc_vo_2.head()
```

```python
plt.figure(figsize=(15,8))
sns.barplot(x=df_loc_vo_2.index,y=df_loc_vo_2['votes'],palette='inferno')
plt.xticks(rotation=90)
```
