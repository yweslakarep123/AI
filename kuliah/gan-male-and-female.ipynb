```python
import os
import numpy as np
import tensorflow as tf
from keras.preprocessing import image
from keras.applications.vgg16 import VGG16, preprocess_input
from sklearn.ensemble import IsolationForest
from shutil import move, copy2
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
from torch.utils import data
from torchvision import datasets, transforms
import matplotlib.pyplot as plt
import pickle as pkl
import cv2
from glob import glob
import random

#  Check for GPU
gpus = tf.config.list_physical_devices('GPU')
if gpus:
    print(f" GPU found: {gpus[0].name}")
else:
    print(" No GPU found. Using CPU.")

# Set paths
dataset_path = '/kaggle/input/male-and-female-faces-dataset/Male and Female face dataset'
output_path = '/kaggle/working/'
os.makedirs(output_path, exist_ok=True)

# Load VGG16 model for feature extraction
model = VGG16(weights='imagenet', include_top=False, pooling='avg')

def load_images_from_folder(folder):
    features = []
    paths = []
    for label in ['Male Faces', 'Female Faces']:
        subfolder = os.path.join(folder, label)
        for fname in os.listdir(subfolder):
            fpath = os.path.join(subfolder, fname)
            try:
                img = image.load_img(fpath, target_size=(224, 224))
                x = image.img_to_array(img)
                x = np.expand_dims(x, axis=0)
                x = preprocess_input(x)
                feat = model.predict(x, verbose=0)[0]
                features.append(feat)
                paths.append(fpath)
            except Exception as e:
                print(f"Could not process {fpath}: {e}")
    return np.array(features), paths

# Extract features and detect outliers
features, image_paths = load_images_from_folder(dataset_path)
iso_forest = IsolationForest(contamination=0.1, random_state=42)  # Increased contamination
outliers = iso_forest.fit_predict(features)

# Move outliers to a separate folder
for path, is_outlier in zip(image_paths, outliers):
    if is_outlier == -1:
        label = 'male' if 'Male' in path else 'female'
        dest_folder = os.path.join(output_path, f'outliers_{label}')
        os.makedirs(dest_folder, exist_ok=True)
        copy2(path, os.path.join(dest_folder, os.path.basename(path)))
        print(f"Removed outlier: {path}")

# Face extraction code
cascade_file = 'haarcascade_frontalface_alt.xml'
cascade_dir = 'haarcascades'
cascade_path = os.path.join(cascade_dir, cascade_file)

if not os.path.exists(cascade_dir):
    os.makedirs(cascade_dir)

if not os.path.exists(cascade_path):
    print(f"Downloading {cascade_file}...")
    cascade_url = f'https://raw.githubusercontent.com/opencv/opencv/master/data/haarcascades/{cascade_file}'
    os.system(f'wget -O {cascade_path} {cascade_url}')

# Get all image files
image_extensions = ['*.jpg', '*.jpeg', '*.png', '*.bmp', '*.tiff', '*.tif', '*.webp',
                   '*.JPG', '*.JPEG', '*.PNG', '*.BMP', '*.TIFF', '*.TIF', '*.WEBP']
male_photos = []
female_photos = []
for ext in image_extensions:
    male_photos.extend(glob(os.path.join(dataset_path, 'Male Faces', ext)))
    female_photos.extend(glob(os.path.join(dataset_path, 'Female Faces', ext)))

photos_with_labels = []
for photo in male_photos:
    photos_with_labels.append((photo, 'male'))
for photo in female_photos:
    photos_with_labels.append((photo, 'female'))

print(f"Found {len(male_photos)} male photos and {len(female_photos)} female photos")

# Create output directories
os.makedirs('extracted_faces/male', exist_ok=True)
os.makedirs('extracted_faces/female', exist_ok=True)

cascade = cv2.CascadeClassifier(cascade_path)
if cascade.empty():
    print(f"Error loading cascade classifier from {cascade_path}")
else:
    print(f"Successfully loaded cascade classifier from {cascade_path}")

# Process each image
face_count = 0
for file_path, gender in photos_with_labels:
    img = cv2.imread(file_path)
    if img is None:
        continue
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    faces = cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5, minSize=(30, 30))
    
    if len(faces) > 0:
        for i, (x, y, w, h) in enumerate(faces):
            face = img[y:y + h, x:x + w]
            # Resize face to consistent size
            face_resized = cv2.resize(face, (64, 64))
            output_filename = f'extracted_faces/{gender}/face_{gender}_{face_count}_{i}.jpg'
            cv2.imwrite(output_filename, face_resized)
            face_count += 1

print(f"Extracted {face_count} faces total")

# Data loader
def get_loaders(batch_size, img_size, data_dir):
    transform = transforms.Compose([
        transforms.Resize(img_size),
        transforms.CenterCrop(img_size),
        transforms.ToTensor(),
        transforms.Normalize((0.5, 0.5, 0.5), (0.5, 0.5, 0.5))
    ])
    d = datasets.ImageFolder(data_dir, transform=transform)
    data_loader = data.DataLoader(d, batch_size=batch_size, shuffle=True, drop_last=True)
    return data_loader

# IMPROVED DISCRIMINATOR - Less aggressive
class Discriminator(nn.Module):
    def __init__(self, conv_dim):
        super(Discriminator, self).__init__()
        self.conv_dim = conv_dim
        
        # Remove batch norm from first layer and reduce aggressiveness
        self.conv1 = nn.Conv2d(3, self.conv_dim, 4, stride=2, padding=1, bias=False)
        self.conv2 = nn.Conv2d(self.conv_dim, self.conv_dim*2, 4, stride=2, padding=1, bias=False)
        self.conv3 = nn.Conv2d(self.conv_dim*2, self.conv_dim*4, 4, stride=2, padding=1, bias=False)
        self.conv4 = nn.Conv2d(self.conv_dim*4, self.conv_dim*8, 4, stride=2, padding=1, bias=False)
        
        # Add dropout layers
        self.dropout = nn.Dropout(0.3)
        self.fc1 = nn.Linear(self.conv_dim*8*4*4, 1)
        
        # Batch norm for layers 2-4 only
        self.batchn2 = nn.BatchNorm2d(self.conv_dim*2)
        self.batchn3 = nn.BatchNorm2d(self.conv_dim*4)
        self.batchn4 = nn.BatchNorm2d(self.conv_dim*8)
        
    def forward(self, x):
        # No batch norm on first layer
        x = F.leaky_relu(self.conv1(x), 0.2)
        x = self.dropout(x)
        
        x = F.leaky_relu(self.batchn2(self.conv2(x)), 0.2)
        x = self.dropout(x)
        
        x = F.leaky_relu(self.batchn3(self.conv3(x)), 0.2)
        x = self.dropout(x)
        
        x = F.leaky_relu(self.batchn4(self.conv4(x)), 0.2)
        
        x = x.view(-1, self.conv_dim*8*4*4)
        x = self.fc1(x)
        return x

# IMPROVED GENERATOR
class Generator(nn.Module):
    def __init__(self, z_size, conv_dim):
        super(Generator, self).__init__()
        self.z_size = z_size
        self.conv_dim = conv_dim
        
        self.fc = nn.Linear(self.z_size, self.conv_dim*8*4*4)
        self.tconv1 = nn.ConvTranspose2d(self.conv_dim*8, self.conv_dim*4, 4, stride=2, padding=1, bias=False)
        self.tconv2 = nn.ConvTranspose2d(self.conv_dim*4, self.conv_dim*2, 4, stride=2, padding=1, bias=False)
        self.tconv3 = nn.ConvTranspose2d(self.conv_dim*2, self.conv_dim, 4, stride=2, padding=1, bias=False)
        self.tconv4 = nn.ConvTranspose2d(self.conv_dim, 3, 4, stride=2, padding=1, bias=False)
        
        self.batchn1 = nn.BatchNorm2d(self.conv_dim*4)
        self.batchn2 = nn.BatchNorm2d(self.conv_dim*2)
        self.batchn3 = nn.BatchNorm2d(self.conv_dim)
        
    def forward(self, x):
        x = self.fc(x)
        x = x.view(-1, self.conv_dim*8, 4, 4)
        
        x = F.relu(self.batchn1(self.tconv1(x)))
        x = F.relu(self.batchn2(self.tconv2(x)))
        x = F.relu(self.batchn3(self.tconv3(x)))
        x = torch.tanh(self.tconv4(x))
        return x

def weights_init_normal(m):
    classname = m.__class__.__name__
    if "Linear" in classname:
        torch.nn.init.normal_(m.weight, 0.0, 0.02)
        if m.bias is not None:
            m.bias.data.fill_(0.0)
    elif "Conv" in classname or "BatchNorm2d" in classname:
        torch.nn.init.normal_(m.weight, 0.0, 0.02)

def build_network(d_conv_dim, g_conv_dim, z_size):
    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    print(f"Using device: {device}")
    
    D = Discriminator(d_conv_dim)
    G = Generator(z_size=z_size, conv_dim=g_conv_dim)
    
    D.apply(weights_init_normal)
    G.apply(weights_init_normal)
    
    D = D.to(device)
    G = G.to(device)
    
    return D, G

# IMPROVED LOSS FUNCTIONS
def real_loss(D_out, smooth=True):
    batch_size = D_out.size(0)
    device = D_out.device
    
    if smooth:
        # Use label smoothing to prevent overconfidence
        labels = torch.ones(batch_size, device=device) * 0.9
    else:
        labels = torch.ones(batch_size, device=device)
    
    criterion = nn.BCEWithLogitsLoss()
    loss = criterion(D_out.squeeze(), labels)
    return loss

def fake_loss(D_out, smooth=True):
    batch_size = D_out.size(0)
    device = D_out.device
    
    if smooth:
        # Add noise to fake labels too
        labels = torch.zeros(batch_size, device=device) + 0.1 * torch.rand(batch_size, device=device)
    else:
        labels = torch.zeros(batch_size, device=device)
    
    criterion = nn.BCEWithLogitsLoss()
    loss = criterion(D_out.squeeze(), labels)
    return loss

# IMPROVED TRAINING FUNCTION
def train(D, G, n_epochs, data_loader, print_every=50):
    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    print(f"Using device: {device}")
    
    D = D.to(device)
    G = G.to(device)
    
    # IMPROVED OPTIMIZERS with different learning rates
    d_lr = 0.0001  # Slower learning for discriminator
    g_lr = 0.0002  # Faster learning for generator
    beta_1 = 0.5
    beta_2 = 0.999
    
    d_optimizer = optim.Adam(D.parameters(), lr=d_lr, betas=(beta_1, beta_2))
    g_optimizer = optim.Adam(G.parameters(), lr=g_lr, betas=(beta_1, beta_2))
    
    # Learning rate schedulers
    d_scheduler = optim.lr_scheduler.ExponentialLR(d_optimizer, gamma=0.99)
    g_scheduler = optim.lr_scheduler.ExponentialLR(g_optimizer, gamma=0.99)
    
    samples = []
    losses = []
    sample_size = 16
    z_size = 100
    
    # Fixed noise for consistent sample generation
    fixed_z = torch.randn(sample_size, z_size, device=device)
    
    for epoch in range(n_epochs):
        for batch_i, (real_images, _) in enumerate(data_loader):
            batch_size = real_images.size(0)
            real_images = real_images.to(device)
            
            # TRAIN DISCRIMINATOR (less frequently to balance training)
            if batch_i % 2 == 0:  # Train discriminator every other batch
                d_optimizer.zero_grad()
                
                # Real images
                real_output = D(real_images)
                d_real_loss = real_loss(real_output, smooth=True)
                
                # Fake images
                z = torch.randn(batch_size, z_size, device=device)
                fake_images = G(z).detach()  # Detach to avoid backprop through G
                fake_output = D(fake_images)
                d_fake_loss = fake_loss(fake_output, smooth=True)
                
                d_loss = d_real_loss + d_fake_loss
                d_loss.backward()
                d_optimizer.step()
            
            # TRAIN GENERATOR
            g_optimizer.zero_grad()
            
            z = torch.randn(batch_size, z_size, device=device)
            fake_images = G(z)
            fake_output = D(fake_images)
            
            # Generator wants discriminator to think fake images are real
            g_loss = real_loss(fake_output, smooth=False)
            g_loss.backward()
            g_optimizer.step()
            
            # Print progress
            if batch_i % print_every == 0:
                if batch_i % 2 == 0:
                    losses.append((d_loss.item(), g_loss.item()))
                    print('Epoch [{:5d}/{:5d}] | d_loss: {:6.4f} | g_loss: {:6.4f}'.format(
                            epoch+1, n_epochs, d_loss.item(), g_loss.item()))
                else:
                    print('Epoch [{:5d}/{:5d}] | d_loss: ------ | g_loss: {:6.4f}'.format(
                            epoch+1, n_epochs, g_loss.item()))
        
        # Update learning rates
        d_scheduler.step()
        g_scheduler.step()
        
        # Generate samples
        G.eval()
        with torch.no_grad():
            sample_z = G(fixed_z)
            sample_z = sample_z.cpu()
            samples.append(sample_z)
        G.train()
        
        # Print epoch summary
        print(f"Epoch {epoch+1}/{n_epochs} completed")
    
    # Save samples
    with open('GAN_samples.pkl', 'wb') as f:
        pkl.dump(samples, f)
    
    return losses

# MAIN EXECUTION
extracted_faces_dir = '/kaggle/working/extracted_faces'
batch_size = 32
img_size = 64
data_loader = get_loaders(batch_size, img_size, extracted_faces_dir)

print(f"DataLoader created!")
print(f"Number of classes: {len(data_loader.dataset.classes)}")
print(f"Classes: {data_loader.dataset.classes}")
print(f"Total samples: {len(data_loader.dataset)}")

# Build networks
d_conv_dim = 32
g_conv_dim = 64  # Reduced generator complexity
z_size = 100

D, G = build_network(d_conv_dim, g_conv_dim, z_size)

# Train the model
n_epochs = 20
losses = train(D, G, n_epochs, data_loader)

# Plot losses
fig, ax = plt.subplots(figsize=(10, 6))
losses = np.array(losses)
plt.plot(losses.T[0], label='Discriminator', alpha=0.7)
plt.plot(losses.T[1], label='Generator', alpha=0.7)
plt.title("Training Losses")
plt.xlabel("Iteration")
plt.ylabel("Loss")
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()

# View samples function
def view_samples(epoch, samples):
    fig, axes = plt.subplots(figsize=(16, 4), nrows=2, ncols=8, sharey=True, sharex=True)
    for ax, img in zip(axes.flatten(), samples[epoch]):
        img = img.detach().cpu().numpy()
        img = np.transpose(img, (1, 2, 0))
        img = ((img + 1) * 255 / 2).astype(np.uint8)
        ax.xaxis.set_visible(False)
        ax.yaxis.set_visible(False)
        ax.imshow(img)
    plt.tight_layout()
    plt.savefig('GAN_faces.png', dpi=150, bbox_inches='tight')
    plt.show()

# Load and view final samples
with open('GAN_samples.pkl', 'rb') as f:
    samples = pkl.load(f)

view_samples(-1, samples)
```
