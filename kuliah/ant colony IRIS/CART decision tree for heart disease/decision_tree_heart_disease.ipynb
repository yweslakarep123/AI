<h2>Tampilkan seberapa penting nya kolom di data menggunakan ANOVA F-value untuk memprediksi apakah orang tersebut terkena serangan jantung atau tidak</h2>

```python
#There is no outliears
# Feature selection using SelectKBest with ANOVA F-value
X = data.drop(["target"], axis = 1)
y = data["target"]
selector = SelectKBest(f_classif, k=13)
X_selected = selector.fit_transform(X, y)

selected_features = X.columns[selector.get_support()]
feature_scores = selector.scores_[selector.get_support()]

# Create a DataFrame to store the feature names and their scores
feature_scores_data_set = pd.DataFrame({'Features': selected_features, 'Scores': feature_scores})

# Sort the DataFrame by score in descending order
feature_scores_data_set = feature_scores_data_set.sort_values(by='Scores', ascending=False)

# Plot the feature scores
plt.figure(figsize=(10, 6))
sns.barplot(x='Scores', y='Features', data=feature_scores_data_set, palette='viridis')
plt.title('Feature Scores')
plt.xlabel('Scores')
plt.ylabel('Features')
plt.show()
```

<h2>Buang data di kolom yang tidak penting untuk memprediksi penyakit jantung</h2>

```python
  X = X.drop(["Fasting Blood Sugar (fbs) > 120 mg/dl","Serum Cholestoral (chol) mg/dl",
            "Resting Electrocardiographic Results (restecg)","Resting Blood Pressure (trestbps)"], axis = 1)
```

<h2>Scale data sebelum dimasukkan ke algoritma tree</h2>
<p>StandardScaler() mempunyai rumus (data-mean)/standard deviasi</p>

```python
#Scale all values for good Accuracy
sc = StandardScaler()
col = ['age',
       'sex',
       'Chest Pain Type (CP)',
       'Thal',
       'Exercise Induced Angina (exang)',
       'Slope of the ST Segment (slope)',
       'Number of Major Vessels (ca)',
       'Maximum Heart Rate Achieved (thalach)',
       'ST depression (oldpeak)']
X[col] = sc.fit_transform(X[col])
X.head()
```

<h2>Pecah data menjadi train dan test untuk train dan evaluasi data</h2>
<p>Didalam code 70 persen dari data akan masuk training dan 30 persen akan masuk data test</p>

<p>di dalam code x menyimbulkan kolom yang digunakan untuk memprediksi target yaitu y</p>

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 42)
```

<h2>Import library decision tree dari sklearn yang mengggunakan algoritma CART dan temukan parameter yang bagus untuk data yang kita pakai</h2>

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV

# Membuat model Decision Tree
clf_tree = DecisionTreeClassifier(random_state=42)

# Menentukan parameter grid untuk pencarian
parameters_tree = {
    'criterion': ['gini', 'entropy'],  # atau 'log_loss' jika Anda menggunakan versi baru dari sklearn
    'max_depth': [None, 10, 20, 30, 40, 50],
    'min_samples_split': [2, 10, 20],
    'min_samples_leaf': [1, 5, 10],
}

# Melakukan Grid Search dengan Cross Validation
grid_clf_tree = GridSearchCV(clf_tree, parameters_tree, cv=6, n_jobs=-1)
grid_clf_tree.fit(X_train, y_train)
```

<h2>Ambil parameter yang bagus dan prediksi data test yang kita buat</h2>

```python
best_model_tree=grid_clf_tree.best_estimator_
y_pred_tree=best_model_tree.predict(X_test)
```

<h2>Hitung akurasi, presisi, dan f1 score dari model tree yang kita pakai</h2>

```python
ac_tree = accuracy_score(y_test, y_pred_tree)
print("Accuracy score for model " f'{best_model_tree} : ',ac_tree)
cr_tree = classification_report(y_test, y_pred_tree)
print("classification_report for model " f'{best_model_tree} : \n',cr_tree)
```

<p>Dari output tersebut menunjukkan model yang kita pakai sangat akurat untuk memprediksi data</p>

<h2>Analisa prediksi data menggunakan matrix yang menunujukkan berapa true postive, False negative, true negative, false positive yang di prediksi model</h2>

```python
cm_rnf = confusion_matrix(y_test, y_pred_tree)
fig, ax = plot_confusion_matrix(conf_mat=cm_rnf ,
                                show_absolute=True,
                                colorbar=True,
                                cmap='autumn',
                               class_names = [True , False ],
                               figsize=(5, 3))
plt.title("CM for Heart Model")
plt.show()
```
