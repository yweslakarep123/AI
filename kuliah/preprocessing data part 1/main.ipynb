<h2>PEMBUATAN DATA RANDOM</h2>

```python
# column 1
point1 = abs(np.random.normal(1, 10, 100))
# column 2
point2 = abs(np.random.normal(2, 10, 100))
# column 3
point3 = abs(np.random.normal(3, 10, 100))
# column 4
point4 = abs(np.random.normal(4, 10, 100))

x = np.c_[point1, point2, point3, point4]
```

```python
y = [int(np.random.randint(0, 4)) for i in range(100)]
```

```python
data = pd.DataFrame()

data['variabel1'] = point1
data['variabel2'] = point2
data['variabel3'] = point3
data['variabel4'] = point4
data['output'] = y
```

```python
data.head()
```

<h2>PLOT 1: LINE PLOT SEMUA VARIABEL</h2>

```python
y = np.linspace(0, 1, 100)
plt.subplot(2, 2, 1)
plt.title('Kolom variabel 1')
plt.plot(y, data['variabel1'], color ='r', label ='kolom variabel 1')
plt.legend()

plt.subplot(2, 2, 2)
plt.title('kolom variabel 2')
plt.plot(y, data['variabel1'], color = 'g', label ='kolom variabel 2')
plt.legend()

plt.subplot(2, 2, 3)
plt.title('kolom variabel 3')
plt.plot(y, data['variabel1'], color ='b', label ='kolom variabel 3')
plt.legend()

plt.subplot(2, 2, 4)
plt.title('kolom variabel 4')
plt.plot(y, data['variabel1'], color ='y', label ='kolom variabel 4')
plt.legend()

plt.show()
```

<h2>PLOT 2: SCATTER PLOT SEMUA VARIABEL</h2>

```python
plt.subplot(2, 2, 1)
plt.title('Kolom variabel 1')
plt.scatter(data['output'], data['variabel1'], color ='r', label ='kolom variabel 1')
plt.legend()

plt.subplot(2, 2, 2)
plt.title('kolom variabel 2')
plt.scatter(data['output'], data['variabel1'], color = 'g', label ='kolom variabel 2')
plt.legend()

plt.subplot(2, 2, 3)
plt.title('kolom variabel 3')
plt.scatter(data['output'], data['variabel1'], color ='b', label ='kolom variabel 3')
plt.legend()

plt.subplot(2, 2, 4)
plt.title('kolom variabel 4')
plt.scatter(data['output'], data['variabel1'], color ='y', label ='kolom variabel 4')
plt.legend()

plt.show()
```

<h2>PLOT 3: PIE CHART OUTPUT</h2>
