<h2>contoh 4</h2>

```python
Months = ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Oct','Sep','Nov','Dec']
def SeparateCity(v):
    for mon in Months:
        if (mon in v):
            return v[:v.find(mon)]
pidato_df['City'] = pidato_df.FileName.apply(SeparateCity)
```

```python
def SeparateDate(r):
    return r.FileName[len(r.City):r.FileName.find('.txt')]

pidato_df['Date'] = pidato_df.apply(SeparateDate,axis=1)
pidato_df.Date = pd.to_datetime(pidato_df.Date,format='%b%d_%Y')
```

```python
def extractDMY(r):
    r['Day'] = r.Date.day
    r['Month'] = r.Date.month
    r['Year'] = r.Date.year
    return r
pidato_df = pidato_df.apply(extractDMY,axis=1)
pidato_df.drop(columns=['FileName'],inplace=True)
pidato_df.head()
```

```python
Words = ['vote','tax','campaign','economy']
def FindWordRatio(row):
    total_n_words = len(row.Content.split(' '))
    for w in Words:
        row['r_{}'.format(w)] = row.Content.count(w)/total_n_words
    return row
    
pidato_df = pidato_df.apply(FindWordRatio,axis=1)
pidato_df.head()
```

<h2>Diagnosa outlier untuk setiap atribut.</h2>

```python
select_columns = ['mpg', 'cylinders', 'displacement', 'horsepower', 'weight', 'acceleration', 'model_year']
dfmpg_possibilities = df_mpg.origin.unique()
plt.figure(figsize=(20, 12))  
for i, sc in enumerate(select_columns):
    for poss in dfmpg_possibilities:
        BM = df_mpg.origin == poss
        box_sr[poss] = df_mpg[BM][sc].values
    plt.subplot(2, 4, i+1)  
    plt.boxplot(box_sr, vert=False)
    plt.yticks([1, 2, 3], [1, 2, 3])
    plt.title(sc)
plt.tight_layout()
plt.show()
```

```python
df_mpg.columns
```

```python
def cap_outliers(df, column):
    # Calculate the 25th and 75th percentiles
    percentile25 = df[column].quantile(0.25)
    percentile75 = df[column].quantile(0.75)
    print(f"25th Percentile of {column}:", percentile25)
    print(f"75th Percentile of {column}:", percentile75)

    # Calculate the interquartile range (IQR)
    iqr = percentile75 - percentile25

    # Calculate the upper and lower limits for outliers
    uplimit = percentile75 + 1.5 * iqr
    lolimit = percentile25 - 1.5 * iqr
    print(f"Upper Limit of {column}:", uplimit, "Lower Limit:", lolimit)

    # Cap the outliers
    df[column] = np.where(
        df[column] > uplimit, uplimit,
        np.where(df[column] < lolimit, lolimit, df[column])
    )
    return df

# Apply the outlier capping function to the necessary columns
columns_to_cap = ['horsepower', 'acceleration', 'mpg']
capped_df_mpg = df_mpg.copy()

for col in columns_to_cap:
    capped_df_mpg = cap_outliers(capped_df_mpg, col)

# Display the first few rows of the modified DataFrame
print(capped_df_mpg.head())
```
