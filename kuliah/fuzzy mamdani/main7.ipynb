```python
import numpy as np
# Define the fuzzy sets for homework and exam scores
homework_low = np.array([0, 30, 60])
homework_medium = np.array([30, 70, 100])
homework_high = np.array([70, 100, 100])

exam_low = np.array([0, 30, 60])
exam_medium = np.array([30, 70, 100])
exam_high = np.array([70, 100, 100])

# Define the input values (homework and exam scores)
homework_score = 80 # For example, homework score of 45
exam_score = 75 # For example, exam score of 75
# Calculate the membership values for each input in the fuzzy sets
homework_low_membership = np.interp(homework_score, homework_low, [0, 1, 0])
homework_medium_membership = np.interp(homework_score, homework_medium, [0, 1,0])
homework_high_membership = np.interp(homework_score, homework_high, [0, 1, 0])
exam_low_membership = np.interp(exam_score, exam_low, [0, 1, 0])
exam_medium_membership = np.interp(exam_score, exam_medium, [0, 1, 0])
exam_high_membership = np.interp(exam_score, exam_high, [0, 1, 0])
# Define the fuzzy rules
rules = [
    (min(homework_low_membership, exam_low_membership), "grade_low"),
    (min(homework_medium_membership, exam_low_membership), "grade_medium"),
    (min(homework_high_membership, exam_low_membership), "grade_medium"),
    (min(homework_low_membership, exam_medium_membership), "grade_medium"),
    (min(homework_medium_membership, exam_medium_membership), "grade_medium"),
    (min(homework_high_membership, exam_medium_membership), "grade_high"),
    (min(homework_low_membership, exam_high_membership), "grade_high"),
    (min(homework_medium_membership, exam_high_membership), "grade_high"),
    (min(homework_high_membership, exam_high_membership), "grade_high"),
]
# Aggregate the rule outputs
grades = {
    "grade_low": 0,
    "grade_medium": 0,
    "grade_high": 0,
}

for rule in rules:
    membership, grade = rule
    grades[grade] = max(grades[grade], membership)
# Calculate the centroid defuzzified grade
defuzzified_grade = (grades["grade_low"] * 30 + grades["grade_medium"] * 70 +
grades["grade_high"] * 100) / (grades["grade_low"] + grades["grade_medium"] +
grades["grade_high"])
if defuzzified_grade <= 40:
    output_grade = "Low"
elif defuzzified_grade <= 70:
    output_grade = "Medium"
else:
    output_grade = "High"
# Output the recommended grade
print("Recommended Grade:", output_grade)
```

```python
import numpy as np

# Define the fuzzy sets for homework and exam scores
homework_low = np.array([0, 30, 60])
homework_medium = np.array([30, 70, 100])
homework_high = np.array([70, 100, 100])

exam_low = np.array([0, 30, 60])
exam_medium = np.array([30, 70, 100])
exam_high = np.array([70, 100, 100])

def calculate_membership(score, low, medium, high):
    """
    Calculate membership values for low, medium, and high fuzzy sets.
    """
    low_membership = np.interp(score, low, [0, 1, 0])
    medium_membership = np.interp(score, medium, [0, 1, 0])
    high_membership = np.interp(score, high, [0, 1, 0])
    return low_membership, medium_membership, high_membership

def convert_membership(low, medium, high):
    """
    Convert membership values to a numerical score.
    """
    return 1 * low + 2 * medium + 3 * high

def calculate_grade(scores):
    """
    Calculate the final grade based on homework and exam scores.
    Returns a dictionary with all intermediate and final values.
    """
    hw1, hw2, hw3, uts, uas = scores

    # Calculate membership values for each score
    hw1_memberships = calculate_membership(hw1, homework_low, homework_medium, homework_high)
    hw2_memberships = calculate_membership(hw2, homework_low, homework_medium, homework_high)
    hw3_memberships = calculate_membership(hw3, homework_low, homework_medium, homework_high)
    uts_memberships = calculate_membership(uts, exam_low, exam_medium, exam_high)
    uas_memberships = calculate_membership(uas, exam_low, exam_medium, exam_high)

    # Convert membership values to numerical scores
    homework1_converted = convert_membership(*hw1_memberships)
    homework2_converted = convert_membership(*hw2_memberships)
    homework3_converted = convert_membership(*hw3_memberships)
    uts_converted = convert_membership(*uts_memberships)
    uas_converted = convert_membership(*uas_memberships)

    # Calculate the average result
    result = (homework1_converted + homework2_converted + homework3_converted + uts_converted + uas_converted) / 5

    # Determine the final grade based on the result
    if result < 1.5:
        final_grade = "C"
    elif result < 2.5:
        final_grade = "B"
    else:
        final_grade = "A"

    return {
        "homework1_converted": homework1_converted,
        "homework2_converted": homework2_converted,
        "homework3_converted": homework3_converted,
        "uts_converted": uts_converted,
        "uas_converted": uas_converted,
        "result": result,
        "final_grade": final_grade
    }

# List of students and their scores
students = {
    "Jasmine": [70, 80, 75, 80, 40],
    "Rafik": [90, 80, 80, 77, 90],
    "Mutia": [50, 55, 60, 10, 30],
    "Patrik": [10, 0, 90, 60, 70],
    "Fajar": [65, 35, 100, 70, 72]
}

# Calculate and print grades for each student
for student, scores in students.items():
    grade_info = calculate_grade(scores)
    print(f"--- {student} ---")
    print(f"Homework 1 converted: {grade_info['homework1_converted']:.2f}")
    print(f"Homework 2 converted: {grade_info['homework2_converted']:.2f}")
    print(f"Homework 3 converted: {grade_info['homework3_converted']:.2f}")
    print(f"UTS converted: {grade_info['uts_converted']:.2f}")
    print(f"UAS converted: {grade_info['uas_converted']:.2f}")
    print(f"Average Result: {grade_info['result']:.2f}")
    print(f"Final Grade: {grade_info['final_grade']}\n")
```
